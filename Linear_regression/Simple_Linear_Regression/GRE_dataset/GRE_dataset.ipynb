```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import train_test_split
from statsmodels.api import OLS
import statsmodels.api
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv('Admission_Predict_Ver1.1.csv',index_col='Serial No.')
data
```

```python
data.corr()
```

```python
independent_var = [col for col in data.columns]
independent_var.pop()
independent_var
x = data[independent_var]
y = data["Chance of Admit "]
```

```python
thresh = 10
for i in np.arange(0,len(independent_var)):
    vif = [variance_inflation_factor(x[independent_var].values,ix) for ix in range(x[independent_var].shape[1])]
    maxloc = vif.index(max(vif))
    if max(vif)>thresh:
        print("vif: ",vif)
        print("Dropping \' " + x[independent_var].columns[maxloc] + '\' at index: ' + str(maxloc))
        del independent_var[maxloc]
    else:
        break
        
independent_var
```

```python
x = data[independent_var]
y = data["Chance of Admit "]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=1)
lr = PolynomialFeatures(degree=10)
x_train_poly = lr.fit_transform(x_train)
lm = OLS(y_train,x_train_poly).fit()
lm.summary()
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error 
r2_score(y_train,lm.predict(x_train_poly)),mean_absolute_error(lm.predict(x_train_poly),y_train),np.sqrt(mean_squared_error(lm.predict(x_train_poly),y_train))
```

```python
from statsmodels.graphics.regressionplots import plot_leverage_resid2
fig,ax = plt.subplots(figsize=(18,6))
fig = plot_leverage_resid2(lm,ax=ax)
```

```python
test = lm.outlier_test()
print(test[test["bonf(p)"] < 0.05])
test
```

```python
plt.plot(lm.resid,'o')
plt.title("Residual Plot")
plt.ylabel("Residual")
plt.xlabel("Observation Number")
plt.show()
plt.hist(lm.resid)
```

```python
fig = plt.figure(figsize=(10,5))
fig = statsmodels.api.graphics.plot_partregress_grid(lm,fig=fig)
```
