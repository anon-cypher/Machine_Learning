```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
data_x,data_y = datasets.load_diabetes(return_X_y=True, as_frame=True)
data_x
```

```python
data_y
```

```python
data_x['data_y'] = data_y
data_x.head()
```

```python
data_x.corr()
```

```python
data_x = data_x.drop(["data_y","s3"],axis=1)
```

```python
train_x,test_x,train_y,test_y=train_test_split(data_x,data_y,test_size=0.3,random_state=1)
```

```python
lr = LinearRegression()
lr.fit(train_x,train_y)
```

```python
lr.coef_,lr.intercept_
```

```python
r2_score(test_y,lr.predict(test_x)),mean_squared_error(lr.predict(test_x),test_y)
```

using Ordinary least square Regression

```python
import statsmodels.api as sm 
```

```python
ordls = sm.OLS(train_y,train_x)
model = ordls.fit()
r2_score(test_y,model.predict(test_x)),mean_squared_error(model.predict(test_x),test_y)
```
