```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('train.csv')
data.head(5)
```

```python
sns.countplot(x='Survived',hue='Pclass',data=data)
```

```python
data.isna().sum()
```

```python
emb = pd.get_dummies(data.Embarked,drop_first=True)
pclass = pd.get_dummies(data.Pclass,drop_first=True)
sex = pd.get_dummies(data.Sex,drop_first=True)

data = data.join([emb,pclass,sex])
data
```

```python
data.drop(['PassengerId','Name','Sex','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.corr()
```

```python
sns.heatmap(data)
```

```python
data.dropna(inplace=True)
data.head()
```

```python
x = data.iloc[:,2:]
y = data['Survived']

```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_trans = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_trans,y,test_size=0.3,random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
model = lr.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import f1_score,accuracy_score
```

```python
f1_score(y,model.predict(x_trans)),accuracy_score(y,model.predict(x_trans))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y,model.predict(x_trans))
```
